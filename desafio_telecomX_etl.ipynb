````python
import pandas as pd

# URL correta do arquivo JSON
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/main/TelecomX_Data.json"

# LÃª o arquivo JSON diretamente da internet
df = pd.read_json(url)

# Mostra as primeiras linhas da tabela
df.head()
````

````python
df.head()
````

````python
df.isnull().sum()
````

````python
df.drop_duplicates(inplace=True)
````

````python
# Separar os dados das colunas que contÃªm dicionÃ¡rios

# Expandir a coluna 'customer'
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# Juntar tudo em um novo DataFrame completo
df_final = pd.concat([df[['customerID', 'Churn']], customer_df, phone_df, internet_df, account_df], axis=1)

# Visualizar os dados tratados
df_final.head()
````

````python
df_final.drop_duplicates(inplace=True)
````

````python
df_final.columns = [col.lower().strip().replace(" ", "_") for col in df_final.columns]
````

````python
df_final.to_csv("dados_tratados.csv", index=False)
````
